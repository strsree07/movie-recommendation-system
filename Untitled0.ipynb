```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# 1. Load the Dataset
# Ensure you have the csv file in the same folder
try:
    movies = pd.read_csv('tmdb_5000_movies.csv')
except:
    # Fallback if you don't have the file yet, creating dummy data to test
    data = {
        'title': ['The Matrix', 'Inception', 'Toy Story', 'Interstellar'],
        'overview': [
            'A computer hacker learns from mysterious rebels about the true nature of his reality.',
            'A thief who steals corporate secrets through the use of dream-sharing technology.',
            'A cowboy doll is profoundly threatened and jealous when a new spaceman figure supplants him.',
            'A team of explorers travel through a wormhole in space in an attempt to ensure humanity\'s survival.'
        ]
    }
    movies = pd.DataFrame(data)
    print("Data Loaded Successfully!")

# 2. Preprocessing: Create the TF-IDF Matrix
# This converts the text 'overview' into numbers (vectors) the AI can understand.
tfidf = TfidfVectorizer(stop_words='english')

# Replace NaN with empty strings just in case
movies['overview'] = movies['overview'].fillna('')

# Construct the required TF-IDF matrix by fitting and transforming the data
tfidf_matrix = tfidf.fit_transform(movies['overview'])

# 3. Compute Similarity Score
# We use linear_kernel (faster than cosine_similarity) to calculate how similar every movie is to every other movie
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# 4. Define the Recommendation Function
# Create a reverse mapping of indices and movie titles
indices = pd.Series(movies.index, index=movies['title']).drop_duplicates()

def get_recommendations(title, cosine_sim=cosine_sim):
    # Get the index of the movie that matches the title
    if title not in indices:
        return "Movie not found in database."

    idx = indices[title]

    # Get the pairwise similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 3 most similar movies (ignoring the first one because it's the movie itself)
    sim_scores = sim_scores[1:4]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 3 most similar movies
    return movies['title'].iloc[movie_indices]

# 5. Test It
print("\n--- Recommendation for 'The Matrix' ---")
print(get_recommendations('The Matrix'))
```
